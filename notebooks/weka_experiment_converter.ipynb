# Weka experiment converter to JSON

## Read weka reports

```python
import pandas as pd
TARGET_DATASET = 'TwReyes2013'
df = pd.read_csv('../reports/{}.weka/reports.csv'.format(TARGET_DATASET))
df
```

## Average among each fold

```python
df_avg = df.groupby(['Key_Dataset']).mean()
df_avg.head()
```

## Extract dict form csv row

```python
def extract_name(index):
    return index.split('labeled_matrix-')[-1]

def extract_report_data(row):
    dict_report = dict()
    # Extract dict report
    dict_report['accuracy'] = row['Percent_correct']
    # Per class score
    pos_instances = row['Num_true_positives'] + row['Num_false_negatives']
    neg_instances = row['Num_true_negatives'] + row['Num_false_positives']
    tot_instances = pos_instances + neg_instances
    dict_report['False'] = {
        "precision": ((row['Weighted_avg_IR_precision'] * tot_instances) -
                      (pos_instances * row['IR_precision'])) / neg_instances,
        "recall": ((row['Weighted_avg_IR_recall'] * tot_instances) -
                   (pos_instances * row['IR_recall'])) / neg_instances,
        "f1-score": ((row['Weighted_avg_F_measure'] * tot_instances) -
                     (pos_instances * row['F_measure'])) / neg_instances,
        "support": neg_instances
        
    }
    dict_report['True'] = {
        "precision": row['IR_precision'],
        "recall": row['IR_recall'],
        "f1-score": row['F_measure'],
        "support": pos_instances
    }
    # Weighted average
    dict_report['weighted avg'] = {
        "precision": row['Weighted_avg_IR_precision'],
        "recall": row['Weighted_avg_IR_recall'],
        "f1-score": row['Weighted_avg_F_measure'],
        "support": tot_instances 
    }
    # Build report dict
    dict_data = dict()
    dict_data['report'] = dict_report
    # Confusion matrix
    tn, fn = row['Num_true_negatives'], row['Num_false_negatives']
    fp, tp = row['Num_false_positives'], row['Num_true_positives']
    dict_data['confusion-matrix'] = [[tn, fn], [fp, tp]]
    # Return dict
    return dict_data

def extract_dict(index, row):
    # Build dict
    out_dict = dict()
    out_dict['overall'] = extract_report_data(row)
    out_dict['features'] = extract_name(index)
    out_dict['classifier'] = 'BayesianNetworks'
    # Return dict
    return out_dict
```

## Dump dict to file

### Path

```python
weka_path = '{}.weka'.format(TARGET_DATASET)
# Get project root
path = !pwd
report_path = '{}/{}/{}'.format('/'.join(path[0].split('/')[:-1]), 'reports', weka_path)
report_path
```

```python
import json

def dump_dict_report(report_dict):
    filename = '{}${}.json'.format(report_dict['features'], 'BayesianNetworks')
    path = '{}/{}'.format(report_path, filename)
    with open(path, 'w') as fp:
        json.dump(report_dict, fp, indent=4)
```

## Convert each report

```python
for index, row in df_avg.iterrows():
    report_dict = extract_dict(index, row)
    dump_dict_report(report_dict)
```
